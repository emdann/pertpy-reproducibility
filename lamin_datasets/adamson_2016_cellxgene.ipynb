```python
geo_id = "GSE90546"
doi = "10.1016/j.cell.2016.11.048"
title = "A Multiplexed Single-Cell CRISPR Screening Platform Enables Systematic Dissection of the Unfolded Protein Response"
```

```python
adata = adata[:, validated].copy()
```

```python
adata.var["feature_is_filtered"] = False
```

```python
adata.X = sparse.csr_matrix(adata.X)
```

```python
adata.obs.drop(columns=["disease", "organism"], inplace=True)
```

```python
adata.uns[UnsConstants.TITLE] = "Pilot experiment of 6000 chronic myeloid leukemia (K562) cells carrying 8 distinct GBCs"
```

```python
sc.pp.pca(adata)
sc.pp.neighbors(adata)
sc.tl.umap(adata)
```

```python
adata.obs["tissue_type"] = "cell culture"
```

```python
disease_record = bt.Disease.from_public(ontology_id="MONDO:0011996")
disease_record.save()
```

```python
ct_record = bt.CellType.from_public(ontology_id="CL:0017005")
ct_record.save()
```

```python
adata.obs[ObsConstants.ASSAY_ONTOLOGY_TERM_ID] = assays.ln_10x_3_transcription_profiling.ontology_id
adata.obs[ObsConstants.SEX_ONTOLOGY_TERM_ID] = SexOntologyTermId.UNKNOWN.value
adata.obs[ObsConstants.TISSUE_ONTOLOGY_TERM_ID] = tissues.bone_marrow.ontology_id
adata.obs[ObsConstants.SUSPENSION_TYPE] = suspension_types.cell.name
adata.obs[ObsConstants.DEVELOPMENT_STAGE_ONTOLOGY_TERM_ID] = development_stages.unknown.ontology_id
adata.obs[ObsConstants.DISEASE_ONTOLOGY_TERM_ID] = disease_record.ontology_id
adata.obs[ObsConstants.SELF_REPORTED_ETHNICITY_ONTOLOGY_TERM_ID] = ethnicities.unknown.ontology_id
adata.obs[ObsConstants.CELL_TYPE_ONTOLOGY_TERM_ID] = ct_record.ontology_id
adata.obs[ObsConstants.DONOR_ID] = "NA"
adata.obs[ObsConstants.TREATMENT_TYPE] = "CRISPR"
adata.obs[ObsConstants.PERTURBATION_TYPE] = adata.obs["perturbation"]
```

```python
validator = Validator(adata)
```

```python
validated = validator.validate(organism="human", tissue_type="cell line")
```

```python
validator.register_genes()
validator.register_labels("assay")
validator.register_labels("development_stage")
validator.register_labels("disease")
validator.register_labels('self_reported_ethnicity')
validator.register_labels('sex')
validator.register_labels('suspension_type')
validator.register_labels('tissue_type')
validator.register_labels('tissue')
validator.register_labels('donor_id')
```

```python
artifact = validator.register_artifact(description=title, organism="human", tissue_type="cell line")
```

```python
adata.obs["perturbation"].replace({"*": "unknown"}, inplace=True)
```

```python
is_perturbation = ln.ULabel(name="is_perturbation")
is_perturbation.save()
perturbation_labels = [ln.ULabel(name=cat) for cat in adata.obs["perturbation"].cat.categories.tolist()]
ln.save(perturbation_labels)
is_perturbation.children.add(*perturbation_labels)
ln.Feature(name="perturbation", type="category").save()
```

```python
is_treatment = ln.ULabel(name="is_treatment")
is_treatment.save()
crispr_treamtent = ln.ULabel(name="CRISPR")
crispr_treamtent.save()
is_treatment.children.add(crispr_treamtent)
ln.Feature(name="treatment", type="category").save()
```

```python
artifact.labels.add(ln.ULabel.from_values(adata.obs.perturbation.values), features.perturbation)
artifact.labels.add(ln.ULabel.from_values(adata.obs.perturbation_type.values), features.treatment)
```

```python
collection = validator.register_collection(
    artifact,
    name=title,
    description=doi,
    reference=geo_id, # accession number (e.g. GSE#, E-MTAB#, etc.)
    reference_type="GEO")
```
